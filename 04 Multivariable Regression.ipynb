## Training & Test Dataset Split

```python
prices = data['PRICE']
features = data.drop('PRICE', axis=1)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, prices, test_size=0.2, random_state=10)

# % of the training set
len(X_train)/len(features)
```

```python
# % of test dataset
X_test.shape[0]/features.shape[0]
```

## Miltivariate Regression Regression

```python
from sklearn.linear_model import LinearRegression as LR
```

```python
regr = LR()
regr.fit(X_train, y_train)
```

```python
regr.score(X_test, y_test)
```

```python
print('Intercept', regr.intercept_)
pd.DataFrame(data=regr.coef_, index=X_train.columns, columns=['coef'])
```

```python
print('Training data r-squared:', regr.score(X_train, y_train))
print('Test data r-squared:', regr.score(X_test, y_test))
```

## M Regression with Log Prices

```python
prices = np.log(data['PRICE']) #log prices
features = data.drop('PRICE', axis=1)


X_train2, X_test2, y_train2, y_test2 = train_test_split(features, prices, test_size=0.2, random_state=10)

regr2 = LR()
regr2.fit(X_train2, y_train2)

print('Training data r-squared:', regr2.score(X_train2, y_train2))
print('Test data r-squared:', regr2.score(X_test2, y_test2))

print('Intercept', regr2.intercept_)
pd.DataFrame(data=regr2.coef_, index=X_train2.columns, columns=['coef'])
```

## p values & evaluating coefficients

```python
import statsmodels.api as sn
```

```python
X_incl_const = sn.add_constant(X_train2)

model = sn.OLS(y_train2, X_incl_const)
results = model.fit()

results.params
```

```python
results.pvalues
```

```python
pd.DataFrame({'coef': results.params,'p-values': round(results.pvalues, 3), 'significant': results.pvalues < 0.05})
```

## Testing for Multicollinearity

$$ TAX = \alpha_0 + \alpha_1 RM + \alpha_2 NOX + ... + \alpha _{12} LSTAT $$

$$ VIF_{TAX} = \frac{1}{(1 - R _ {TAX} ^ 2)} $$

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
variance_inflation_factor(exog=X_incl_const.values, exog_idx=1)
```

```python
## find the number of columns? X_incl.const.shape[1] or len(X_incl_const.columns)
for n in range(len(X_incl_const.columns)-1):
    print(f'VIF for {X_incl_const.columns[n+1]} is {variance_inflation_factor(exog=X_incl_const.values, exog_idx=n+1)}')
```

```python
vif=[]

for n in range(len(X_incl_const.columns)):
    vif.append(variance_inflation_factor(exog=X_incl_const.values, exog_idx=n))
print(vif)
```

```python
vif2=[variance_inflation_factor(exog=X_incl_const.values, exog_idx=n) for n in range(len(X_incl_const.columns))]

pd.DataFrame({'coef_name': X_incl_const.columns, 'vif': np.around(vif2, 2), 'vif is bigger than 10?': [vif2[n] > 10 for n in range(14)]})
```
