## Visualising Data - Histograms, Distributions and Bar Charts

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(10,6))
plt.xlabel('Price in 1000s')
plt.ylabel("No. of Houses")
plt.hist(data['PRICE'], bins=50, ec='black', color='#2196f3')
plt.show()
```

```python
import seaborn as sns
sns.set()
```

```python
plt.figure(figsize=(10,6))
sns.distplot(data['PRICE'], bins=50, kde=False, color='#fbc02d')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.xlabel('Average Number Rooms')
plt.ylabel('No. of Houses')
plt.hist(data['RM'], color='green')
plt.show()
```

## Correlation
## $$\rho _(XY) = corr(X,Y)$$
## $$ -1.0 \leq \rho _{XY} \leq +1.0$$

```python
data['PRICE'].corr(data['RM'])
```

```python
data['PRICE'].corr(data['PTRATIO'])
```

```python
data.corr()
```

```python
import numpy as np

mask = np.zeros_like(data.corr())
triangle_indices = np.triu_indices_from(mask)
mask[triangle_indices] = True
mask
```

```python
plt.figure(figsize = (16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
sns.heatmap(data.corr(), mask = mask, cmap='coolwarm', annot = True, annot_kws={"size":14})
sns.set_style('white')
plt.show()
```

```python
sns.set()
nox_dis_corr = round(data['NOX'].corr(data['DIS']), 3)
plt.figure(figsize=(9,6))
plt.title(f'DIS vs NOX (Correlation {nox_dis_corr})', fontsize=16)
plt.scatter(x=data["DIS"], y=data['NOX'], alpha=0.6, s=80, color='indigo')
plt.xlabel('Distance from employment centers', fontsize=14)
plt.ylabel('Nitric Oxide Pollution', fontsize=14)
plt.show()
```

```python
sns.set()
sns.set_context('notebook')
sns.set_style('darkgrid')
sns.jointplot(x=data['DIS'], y=data['NOX'], size=7, color='indigo', joint_kws={'alpha':0.5})
plt.show()
```

```python
sns.set()
sns.set_context('notebook')
sns.set_style('darkgrid')
sns.jointplot(x=data['DIS'], y=data['NOX'], size=7, color='blue', joint_kws={'alpha':1}, kind='hex')
plt.show()
```

```python
sns.set()
sns.set_context('notebook')
sns.set_style('darkgrid')
sns.jointplot(x=data['TAX'], y=data['RAD'], size=7, color='darkred', joint_kws={'alpha':0.5})
plt.show()
```

```python
sns.lmplot(x='TAX', y='RAD', data=data, size=7)
plt.show()
```

```python
sns.set()
rm_tgt_corr = round(data['RM'].corr(data['PRICE']), 3)
plt.figure(figsize=(9,6))
plt.title(f'RM vs PRICE (Correlation {rm_tgt_corr})', fontsize=16)
plt.scatter(x=data["RM"], y=data['PRICE'], alpha=0.6, s=80, color='skyblue')
plt.xlabel('RM - Median No of Rooms', fontsize=14)
plt.ylabel('PRICE - Property Price in 1000s', fontsize=14)
plt.show()
```

```python
sns.lmplot(x='RM', y='PRICE', data=data, size=7)
plt.show()
```
